```python
import os
import langchain
from langchain.text_splitter import TokenTextSplitter
from langchain.embeddings import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings()
text_splitter = TokenTextSplitter(chunk_size=110, chunk_overlap=20)

def encode_text(text):
    op_texts = text_splitter.split_text(text)
    # if len(op_texts) >1:
    #     print(len(op_texts))
    embeds = embeddings.embed_query(text)
    return embeds
```

```python
import pandas as pd
df = pd.read_csv('../data/IMDB-Movie-Data.csv')
df.head()
```

```python
type(df['Description'][0])
```

```python
df['Combined'] = "Title: "+df["Title"] + "\nGenre: " + df["Genre"]+ "\nDescription: " + df["Description"]+ "\nYear: " + str(df["Year"])+ "\nActors: " + df["Actors"]+ "\nDirector: " + df["Director"]
```

```python
df['desc_embedding'] = df['Combined'].apply(encode_text)
```
